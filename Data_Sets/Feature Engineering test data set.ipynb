```python
features_to_drop = ['ct_state_ttl', 'ct_dst_sport_ltm', 'ct_src_ltm','id']
df_test = df_test.drop(features_to_drop, axis=1)
```

Preprocessing train data

```python
from sklearn.preprocessing import LabelEncoder
```

```python
categorical_col=df_test.select_dtypes(include='object').columns
categorical_col
```

```python
df_test=pd.get_dummies(df_test, columns=['proto', 'service'], drop_first=True) # one hot encoding
```

```python
df_test.shape
```

```python
label_encoder = LabelEncoder()
df_test['state'] = label_encoder.fit_transform(df_test['state'])
df_test['attack_cat'] = label_encoder.fit_transform(df_test['attack_cat'])
```

Feature engineering

```python
# Interaction between source and destination load
df_test['load_interaction'] = df_test['sload'] * df_test['dload']
# Total transaction bytes between source and destination
df_test['total_bytes'] = df_test['sbytes'] + df_test['dbytes']
# Packet flow ratio between source and destination
df_test['pkt_flow_ratio'] = df_test['spkts'] / (df_test['dpkts'] + 1)
# Bytes difference and ratio
df_test['bytes_diff'] = df_test['sbytes'] - df_test['dbytes']
df_test['bytes_ratio'] = df_test['sbytes'] / (df_test['dbytes'] + 1)
# TTL difference
df_test['ttl_diff'] = df_test['sttl'] - df_test['dttl']
# Jitter difference and ratio
df_test['jitter_diff'] = df_test['sjit'] - df_test['djit']
df_test['jitter_ratio'] = df_test['sjit'] / (df_test['djit'] + 1)
# Difference between synack and ackdat times
df_test['tcp_time_diff'] = df_test['synack'] - df_test['ackdat']
```
